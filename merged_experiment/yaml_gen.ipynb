```python
import sys
import os
import numpy as np
import pandas as pd
import yaml
import copy
import random
import csv

settings_temp = yaml.load(open('settings_temp.yml'), Loader=yaml.FullLoader)
#print(settings_temp['stimuli']['movie_files'])
#create settings files for each group folder

#load path from csv
group_path =  pd.read_csv('path.csv', header=None, index_col=0).loc['group_path'][1]
all_group_folders=os.listdir(group_path)
```

```python
#create settings files for one group folder
def create_settings_for_one_group(group_folder):
    #900 final videos for one group,create 10 settings files for one group, 90 videos for each settings file
    #shuffle the final videos and create 10 settings files
    final_path = group_path+group_folder+'/final_videos'
    all_final_videos = os.listdir(final_path)
    random.shuffle(all_final_videos)
    #create 10 settings files for one group
    settings_list = []
    for i in range(10):
        settings_list.append(copy.deepcopy(settings_temp))
    #add 90 videos to each settings file
    
    for i in range(len(all_final_videos)):
        settings_list[i%10]['stimuli']['movie_files'].append(final_path+'/'+all_final_videos[i])
    #write 10 settings files to disk
    for i in range(10):
        with open(group_path+group_folder+'/settings'+str(i+1)+'.yml', 'w') as outfile:
            print(group_path+group_folder+'/settings'+str(i+1)+'.yml:')
            print(settings_list[i])
            print(' ')
            yaml.dump(settings_list[i], outfile, default_flow_style=False)
```

```python
def create_settings_for_all_groups():
    for group_folder in all_group_folders:
        if os.path.isdir(group_path+group_folder):
            create_settings_for_one_group(group_folder)
```

```python
create_settings_for_all_groups()
```
